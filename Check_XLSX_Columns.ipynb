# 1. Import Required Libraries
Import pandas for data manipulation and openpyxl for reading Excel files.

```python
import pandas as pd
```

# 2. Load the Excel File
Load the xlsx file using pandas read_excel() function and display basic info.

```python
# Load the Excel file
excel_path = 'Sample data for Result Analytics Suite.xlsx'
df = pd.read_excel(excel_path)
df.info()
df.head()
```

# 3. Define Expected Columns from Code
Define the list of column names that are expected in the code or schema.

```python
# List of expected columns as per the Django model/code
expected_columns = [
    'USN', 'SubjectCode', 'SubjectName', 'Marks', 'SGPA',
    'Category', 'AdmissionQuota', 'StudentName', 'Semester', 'Grade', 'AcademicYear'
]
print('Expected columns:', expected_columns)
```

# 4. Extract Column Names from Excel File
Extract the column names from the loaded DataFrame.

```python
# Extract columns from the DataFrame
excel_columns = list(df.columns)
print('Excel columns:', excel_columns)
```

```python
# Let's look at more rows to find the actual header row
print("First 15 rows of the Excel file:")
for i in range(15):
    print(f"Row {i}: {list(df.iloc[i])[:5]}")  # Show first 5 columns
```

```python
# Let's see all columns in row 8 (the header row)
print("Actual header row (row 8):")
print(df.iloc[8].tolist())
print("\nTotal columns:", len(df.columns))
```

```python
# Check all sheets in the Excel file
import openpyxl
wb = openpyxl.load_workbook('Sample data for Result Analytics Suite.xlsx')
print("Sheet names:", wb.sheetnames)
```

```python
# Check each sheet for its structure
for sheet_name in wb.sheetnames:
    df_sheet = pd.read_excel(excel_path, sheet_name=sheet_name)
    print(f"\n=== Sheet: {sheet_name} ===")
    print(f"Columns: {list(df_sheet.columns)[:8]}")  # First 8 columns
    print(f"Rows: {len(df_sheet)}")
```

```python
# Let's look at Format 1 more closely - find the header row
df1 = pd.read_excel(excel_path, sheet_name='Format 1', header=None)
print("Looking for header row in Format 1:")
for i in range(min(20, len(df1))):
    row_data = df1.iloc[i].dropna().tolist()
    if row_data:
        print(f"Row {i}: {row_data[:6]}")
```

```python
# Let's see all columns in row 9 (header row) of Format 1
print("Full header row (row 9):")
print(df1.iloc[9].tolist())
print("\n\nLet's check if there's marks/SGPA data in any cell:")
# Search for any cell containing 'marks' or 'sgpa'
for i in range(min(50, len(df1))):
    for j, val in enumerate(df1.iloc[i]):
        if isinstance(val, str) and ('marks' in val.lower() or 'sgpa' in val.lower() or 'subject' in val.lower()):
            print(f"Row {i}, Col {j}: {val}")
```

# 5. Compare Columns
Compare the expected columns with the Excel file columns using set operations.

```python
# Compare columns
expected_set = set(expected_columns)
excel_set = set(excel_columns)

missing_in_excel = expected_set - excel_set
extra_in_excel = excel_set - expected_set

print('Missing columns in Excel:', missing_in_excel)
print('Extra columns in Excel:', extra_in_excel)
```

# 6. Identify Missing and Extra Columns
Identify columns that are missing in the Excel file or extra in the Excel file compared to the expected columns.

```python
if not missing_in_excel and not extra_in_excel:
    print('All columns match!')
else:
    if missing_in_excel:
        print('Missing columns:', missing_in_excel)
    if extra_in_excel:
        print('Extra columns:', extra_in_excel)
```
